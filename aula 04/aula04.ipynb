```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('imigrantes_canada.csv')
```

```python
dados.head()
```

```python
dados.tail()
```

```python
dados.info()
```

```python
dados.set_index(['País'],inplace=True)
anos = list(map(str,range(1980,2013)))
```

```python
anos
dados.tail()
```

```python
brasil = dados.loc['Brasil',anos]
```

```python
brasil
```

```python
brasil_dic = {
    'ano': brasil.index.tolist(),
    'imigrantes': brasil.values.tolist()
}
```

```python
df_brasil = pd.DataFrame(brasil_dic)
```

```python
df_brasil.head()
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel("Anos",fontsize=14)
plt.title('Imigração do Brasil para o Canada',fontsize=18)
plt.ylabel("Número de imigrantes",fontsize=14)
plt.show()
```

```python
argentina = dados.loc['Argentina',anos]
argentina
```

```python
argentina_dic = {
    'ano': argentina.index.tolist(),
    'imigrantes': argentina.values.tolist()
}
```

```python
df_argentina = pd.DataFrame(argentina_dic)
```

```python
df_argentina.head()
```

```python
plt.plot(df_argentina['ano'],df_argentina['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel("Anos",fontsize=14)
plt.title('Imigração da Argentina para o Canada',fontsize=18)
plt.ylabel("Número de imigrantes",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],color='red',label='Brasil')
plt.plot(df_argentina['ano'],df_argentina['imigrantes'],color='blue',label='Argentina')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel("Anos",fontsize=14)
plt.title('Imigração da Argentina e Brasil para o Canada',fontsize=14)
plt.ylabel("Número de imigrantes",fontsize=14)
plt.legend()
#plt.savefig('Dados imigração.jpg')
plt.show()
```

```python
plt.bar(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.title('Imigração do Brasil para o Canada',fontsize=14)
plt.show()
```

```python
df_comparação = dados.loc[['Brasil','Argentina'],anos]
df_comparação
```

```python
df_comparação = df_comparação.T
df_comparação['T'] = df_comparação['Brasil'] + df_comparação['Argentina']
```

```python
plt.figure(figsize=(10,10))
plt.plot(df_comparação['Brasil'],label='Brasil',color='green')
plt.plot(df_comparação['Argentina'],label='Argentina',color='blue')
plt.plot(df_comparação['T'],label='Total',color='red')
plt.legend()
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.title('Imigração do Brasil e \n Argentina para o Canada',fontsize=20)
plt.ylabel("Número de imigrantes",fontsize=14)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=3,color='green',label='brasil')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_tick_params(labelsize= 12)
ax.yaxis.set_tick_params(labelsize= 12)
ax.set_xlabel('Anos',fontsize=12)
ax.set_ylabel('Número de imigrantes',fontsize=12)
ax.set_title("imigrantes do Brasil para o Canadá",fontsize=18)
ax.legend(loc='best')
ax.grid()
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'])
axs[0].set_title('Imigrantes do Brasil para o Canadá de \n 1980 - 2013',fontsize=18)
axs[0].set_xlabel('Anos',fontsize=12)
axs[0].set_ylabel('Anos',fontsize=12)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box plot da imigração do Brasil \n para o Canadá',fontsize=18)
axs[1].set_xlabel('Brasil',fontsize=12)
axs[1].set_ylabel('Número de imigrantes',fontsize=12)
axs[1].grid()
plt.show()
```
