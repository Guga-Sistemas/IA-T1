```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) #conjunto de dados
nivel = np.array([50,60,70,45,80,85])#conjunto de dados

modelo = LinearRegression() #armazena a regressão linear
modelo.fit(vazao, nivel) # Treina o modelo de regressão linear
```

```python
vazao_teste = np.array([10,20,30,5,35,40]).reshape(-1,1)
#vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1)
nivel_previsto = modelo.predict(vazao_teste)

for i in range(len(vazao_teste)):
    print(f"Vazão: {vazao_teste[i][0]}m3/min - Nível previsto: {nivel_previsto[i]} metros")
plt.scatter(vazao,nivel,color='blue', label='Dados de Treinamento')
plt.plot(vazao_teste,nivel_previsto,color='red',label='Dados de teste')
plt.xlabel('Vazão de Água(m3/min)')
plt.ylabel('Nível do Tanque (metro)')
plt.title('Relação dentre Vazão de Água e Nível do Tanque') 
plt.legend()
plt.grid(True)
plt.show()   
```

```python
Quantidade = np.array([0,100,200,300,400,500,600,700]).reshape(-1,1) #conjunto de dados
Temperatura = np.array([0,30,35,40,45,60,100,150])#conjunto de dados

modelo2 = LinearRegression() #armazena a regressão linear
modelo2.fit(Quantidade, Temperatura) # Treina o modelo de regressão linear
```

```python
Quantidade_teste = np.array([0,50,100,150,200,250,700]).reshape(-1,1)

Temperatura_teste = modelo2.predict(Quantidade_teste)

for i in range(len(Quantidade_teste)):
    print(f"Quantidade: {Quantidade_teste[i][0]} - Temperatura prevista: {Temperatura_teste[i]} Graus")
plt.scatter(Quantidade,Temperatura,color='blue', label='Dados de Treinamento')
plt.plot(Quantidade_teste,Temperatura_teste,color='red',label='Dados de teste')
plt.xlabel('Quantidade de Calor')
plt.ylabel('Temperatura do forno')
plt.title('Relação entre Temperatura do forno e Quantidade de Calor') 
plt.legend()
plt.grid(True)
plt.show()   
```

```python
Area = np.array([50,100,120,300]).reshape(-1,1) #conjunto de dados
Valor_imov = np.array([180000,300000,375000,600000])#conjunto de dados

modelo3 = LinearRegression() #armazena a regressão linear
modelo3.fit(Area, Valor_imov) # Treina o modelo de regressão linear
```
